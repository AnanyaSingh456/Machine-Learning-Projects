Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = pd.read_csv("/content/boston-house-prices.csv")

# renaming the 'MEDV' column name to 'Price'
house_price_dataset.rename(columns={'MEDV':'Price'}, inplace=True)
house_price_dataset.head()
```

```python
# checking the number of rows and columns
house_price_dataset.shape
```

```python
# check for missing values
house_price_dataset.isnull().sum()
```

```python
# statistical measures
house_price_dataset.describe()
```

Understanding the correlation between various features in the dataset


*   Positive Correlation
*   Negative Correlation



```python
correlation = house_price_dataset.corr()
correlation.head()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and target

```python
X = house_price_dataset.drop(['Price'], axis=1)
Y = house_price_dataset['Price']
```

```python
print(X)
print(Y)
```

Splitting the data into Training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training



XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model
model.fit(X_train, Y_train)
```

Model Evaluation

Prediction on Training Data

```python
Y_train_pred = model.predict(X_train)
Y_train_pred
```

```python
# R squared error
error_score = metrics.r2_score(Y_train, Y_train_pred)
print("R squared error : ", error_score)

# Mean Absolute Error
mae = metrics.mean_absolute_error(Y_train, Y_train_pred)
print("Mean Absolute Error : ", mae)
```

Visualizing the Actual Prices and Predicted Prices

```python
plt.scatter(Y_train, Y_train_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
```

Prediction on Test Data

```python
Y_test_pred = model.predict(X_test)
Y_test_pred
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, Y_test_pred)
print("R squared error : ", error_score)

# Mean Absolute Error
mae = metrics.mean_absolute_error(Y_test, Y_test_pred)
print("Mean Absolute Error : ", mae)
```

Visualizing the Actual Prices and Predicted Priced


```python
plt.scatter(Y_test, Y_test_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
```
